### **Webscraping Time**

**Men's Swimming Team**

```
#Brooklyn College

page = requests.get(BrooklynSwimmingMen)
soup = BeautifulSoup(page.content, 'html.parser')
NameList = soup.find_all('a', class_='sidearm-roster-player-name')
HeightList = soup.find_all('span', class_='sidearm-roster-player-height')

Names = []
Heights = []
HeightsAdjusted = []


for name in NameList:
  Names.append(name.get_text().strip('\n').replace('\n', ' '))

for height in HeightList:
  Heights.append(height.get_text().strip('"'))

#I create a formula to convert the heights of players from ft to cm for more accurate calculations and analysis
for i in range(0,len(Heights)):
  convertion = (float(Heights[i][0]) * 30.48) +  (float(Heights[i][2]) * 2.54)
  HeightsAdjusted.append(round(convertion,2))

Names = Names[:12]
HeightsAdjusted = HeightsAdjusted[:12]

BrooklynSwimmingMendf = pd.DataFrame()
BrooklynSwimmingMendf['Athletes Name'] = Names
BrooklynSwimmingMendf['Heights'] = HeightsAdjusted
```

```
#Question 5: Find the average height in each of the 4 dataframes (so you should have 4 averages in total)

def Average_Height(team):
  return team['Heights'].mean()

print("The average height for men's swimming team is:", str((round(Average_Height(MenSwimmingTeamdf),2))) + 'cm')
print('\n')
print("The average height for women's swimming team is:", str(round(Average_Height(WomenSwimmingTeamdf),2)) + 'cm')
print('\n')
print("The average height for men's volleyball team is:", str(round(Average_Height(MenVolleyTeamdf),2)) + 'cm')
print('\n')
print("The average height for women's volleyball team is:", str(round(Average_Height(WomenVolleyTeamdf),2)) + 'cm')
```
